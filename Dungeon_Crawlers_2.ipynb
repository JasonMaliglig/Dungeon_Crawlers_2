```python
class Weapons:
    def __init__(self, name, damage, rarity="common"):
        self.name = name
        self.damage = damage
        self.rarity = rarity
        self.boost_duration = 0
        self.poison_duration = 0

    def do_damage(self, target):
        return self.damage
    
    def damage_boost(self, amount, duration):
        if self.boost_duration <= 0:
            self.damage += amount
            self.boost_duration = duration
        else:
            print("Damage boost is already active.")
    
    def update_boost_duration(self):
        if self.boost_duration > 0:
            self.boost_duration -= 1
            if self.boost_duration == 0:
                self.damage -= amount

    def poison_boost(self, amount, duration):
        if self.poison_boost_duration <= 0:
            self.damage += amount
            self.boost_duration = duration
        else:
            print("Damage boost is already active.")

class Sword(Weapons):
    def __init__(self, rarity="common"):
        super().__init__("Sword", 10, rarity)


class Staff(Weapons):
    def __init__(self, rarity="common"):
        super().__init__("Staff", 10, rarity)


class PoisonedKnife(Weapons):
    def __init__(self, rarity="common"):
        super().__init__("Poisoned Knife", 10, rarity)


class Bow(Weapons):
    def __init__(self, rarity="common"):
        super().__init__("Bow", 10, rarity)


# Potion Classes
class Potions:
    def __init__(self, rarity="common"):
        self.rarity = rarity
        self.damage_boost_duration = 2
        self.poison_damage_duration = 5


class HealthPotion(Potions):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Health Potion"
        self.healing_amount = 50


class ManaPotion(Potions):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Mana Potion"
        self.mana_amount = 50


class DamagePotion(Potions):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Damage Potion"
        self.damage_boost = 5


class PoisonPotion(Potions):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Poison Potion"
        self.poison_damage = 2

# Armor Class
class Armor:
    def __init__(self, rarity="common"):
        self.rarity = rarity
        self.cooldown = 0
        self.cooldown_duration = 2

class Shield(Armor):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Shield"
        self.block_amount = 10
        
class Cloak(Armor):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Cloak"
        self.block_amount = 5
        
class Chainmail(Armor):
    def __init__(self, rarity="common"):
        super().__init__(rarity)
        self.name = "Chainmail"
        self.block_amount = 5
        
# Player Classes
class PlayerType:
    def __init__(self, name):
        self.name = name
        self.inventory = []
        self.health = 100
        self.mana = 100
        self.stamina = 100

    def view_inventory(self):
        print(f"Inventory:")
        for item in self.inventory:
            if isinstance(item, Armor):
                print(f"{item.rarity.capitalize()} {item.name}: Mitigation {item.block_amount}")
            elif isinstance(item, Weapons):
                print(f"{item.rarity.capitalize()} {item.name}: Damage {item.damage}")
                
    def reduce_stamina(self, amount=5):
        self.stamina -= amount
        if self.stamina < 0:
            self.stamina = 0

    def is_stamina_zero(self):
        return self.stamina == 0

    def can_take_action(self):
        return self.stamina > 0
                
    def upgrade_inventory(self):
        for item in self.inventory:
            if isinstance(item, Armor):
                if item.rarity == "common":
                    item.rarity = "uncommon"
                    item.block_amount += 2
                elif item.rarity == "uncommon":
                    item.rarity = "rare"
                    item.block_amount += 2
                elif item.rarity == "rare":
                    item.rarity = "epic"
                    item.block_amount += 2
                elif item.rarity == "epic":
                    item.rarity = "legendary"
                    item.block_amount += 2
            elif isinstance(item, Weapons):
                if item.rarity == "common":
                    item.rarity = "uncommon"
                    item.damage += 2
                elif item.rarity == "uncommon":
                    item.rarity = "rare"
                    item.damage += 2
                elif item.rarity == "rare":
                    item.rarity = "epic"
                    item.damage += 2
                elif item.rarity == "epic":
                    item.rarity = "legendary"
                    item.damage += 2
            elif isinstance(item, Potions):
                if isinstance(item, HealthPotion):
                    if item.rarity == "common":
                        item.rarity = "uncommon"
                        item.healing_amount += 5
                    elif item.rarity == "uncommon":
                        item.rarity = "rare"
                        item.healing_amount += 5
                    elif item.rarity == "rare":
                        item.rarity = "epic"
                        item.healing_amount += 5
                    elif item.rarity == "epic":
                        item.rarity = "legendary"
                        item.healing_amount += 5
                elif isinstance(item, ManaPotion):
                    if item.rarity == "common":
                        item.rarity = "uncommon"
                        item.mana_amount += 5
                    elif item.rarity == "uncommon":
                        item.rarity = "rare"
                        item.mana_amount += 5
                    elif item.rarity == "rare":
                        item.rarity = "epic"
                        item.mana_amount += 5
                    elif item.rarity == "epic":
                        item.rarity = "legendary"
                        item.mana_amount += 5
                elif isinstance(item, DamagePotion):
                    if item.rarity == "common":
                        item.rarity = "uncommon"
                        item.damage_boost += 5
                    elif item.rarity == "uncommon":
                        item.rarity = "rare"
                        item.damage_boost += 5
                    elif item.rarity == "rare":
                        item.rarity = "epic"
                        item.damage_boost += 5
                    elif item.rarity == "epic":
                        item.rarity = "legendary"
                        item.damage_boost += 5
                elif isinstance(item, PoisonPotion):
                    if item.rarity == "common":
                        item.rarity = "uncommon"
                        item.poison_damage += 5
                    elif item.rarity == "uncommon":
                        item.rarity = "rare"
                        item.poison_damage += 5
                    elif item.rarity == "rare":
                        item.rarity = "epic"
                        item.poison_damage += 5
                    elif item.rarity == "epic":
                        item.rarity = "legendary"
                        item.poison_damage += 5

    def take_damage(self, amount):
        self.health -= amount
        if self.health <= 0:
            print(f"You have been defeated!")
            return True
        return False

    def heal(self, amount):
        self.health += amount
        if self.health > 100:
            self.health = 100

    def restore_mana(self, amount):
        self.mana += amount
        if self.mana > 100:
            self.mana = 100


class Knight(PlayerType):
    def __init__(self):
        super().__init__("Knight")
        self.inventory.extend([Sword(), Shield(), HealthPotion(), DamagePotion()])


class Wizard(PlayerType):
    def __init__(self):
        super().__init__("Wizard")
        self.inventory.extend([Staff(), Cloak(), HealthPotion(), DamagePotion()])


class Archer(PlayerType):
    def __init__(self):
        super().__init__("Archer")
        self.inventory.extend([Bow(), Cloak(), HealthPotion(), DamagePotion()])


class Rogue(PlayerType):
    def __init__(self):
        super().__init__("Rogue")
        self.inventory.extend([PoisonedKnife(), Chainmail(), HealthPotion(), PoisonPotion()])

# Enemy Classes
class enemyType:
    def __init__(self, name, health, damage):
        self.name = name
        self.health = health
        self.damage = damage
        self.inventory = []
        self.dazed_duration = 0

    def take_damage(self, amount):
        self.health -= amount
        if self.health <= 0:
            print(f"{self.name} has been defeated!")
            return True
        return False
    
    def deal_damage(self, target):
        if self.dazed_duration <= 0:
            target.take_damage(self.damage)
            print(f"{self.name} attacks you. \033[1mYou take {self.damage} damage.\033[0m")
        else:
            print(f"{self.name} is dazed")
            self.dazed_duration -= 1
            
            
class Spider(enemyType):
    def __init__(self):
        super().__init__("Huge Spider", 20, 5)


class Goblin(enemyType):
    def __init__(self):
        super().__init__("Goblin holding two daggers", 30, 5)


class Cyclopse(enemyType):
    def __init__(self):
        super().__init__("Cyclopse holding a club", 50, 10)


class cursedKnight(enemyType):
    def __init__(self):
        super().__init__("Cursed Knight", 40, 10)


class cursedArcher(enemyType):
    def __init__(self):
        super().__init__("Cursed Archer", 35, 5)


class cursedJester(enemyType):
    def __init__(self):
        super().__init__("Cursed Jester", 25, 5)


class Warewolf(enemyType):
    def __init__(self):
        super().__init__("Warewolf", 50, 5)


class Ewoks(enemyType):
    def __init__(self):
        super().__init__("5 Ewoks with spears running across the foliage", 25, 5)


class Ent(enemyType):
    def __init__(self):
        super().__init__("10ft tall Ent holding a club", 60, 10)

# Path Classes
class pathType:
    def __init__(self, name):
        self.name = name
        self.inventory = []

    def view_path(self):
        print(f"You see the following:")
        for item in self.inventory:
            print(item.name)


class undergroundCavern(pathType):
    def __init__(self):
        super().__init__("Underground Cavern")
        self.inventory.extend([Spider(), Goblin(), Cyclopse()])


class abandonedCastle(pathType):
    def __init__(self):
        super().__init__("Abandoned Castle")
        self.inventory.extend([cursedKnight(), cursedArcher(), cursedJester()])


class cursedForest(pathType):
    def __init__(self):
        super().__init__("Cursed Forest")
        self.inventory.extend([Warewolf(), Ewoks(), Ent()])

#Display Functions and User Interface
def display_classes_with_inventory():
    print("\033[1mAvailable Player Classes:\033[0m", end="\n\n")
    classes = [Knight(), Wizard(), Archer(), Rogue()]
    for index, player_class in enumerate(classes, start=1):
        print(f"--- {index} ---")
        print(f"\033[1m{player_class.name}\033[0m")
        player_class.view_inventory()
        print()


def display_chosen_class(chosen_player):
    print("\033[1mYour Class:\033[0m")
    print(f"{chosen_player.name}")
    print()
    chosen_player.view_inventory()
    print()


def display_path_choice():
    print("\033[1mChoose your path:\033[0m")
    print("(1) Left: Underground cavern")
    print("(2) Middle: Abandoned Castle")
    print("(3) Right: Cursed Forest")
    print()
    
def display_chosen_path(choose_path_object, enemy):
    print("You chose the path:", choose_path_object.name)
    print()
    print("You see three enemies in your way:")
    for enemy in enemy:
        print(f"\033[1m{enemy.name}\033[0m (Health: {enemy.health})", end=" ") 
    print()
    


def display_action_choice(chosen_player):
    print()
    print(f"\033[1mAvailable Actions:\033[0m")
    print("(1) Attack")
    if chosen_player.inventory[1].cooldown <= 0:
        print("(2) Block")
    else:
        print(f"(2) Block (On cooldown for {chosen_player.inventory[1].cooldown} more rounds)")
    print("(3) Use Potion")
    print("(4) Craft Potion")
    print("(5) View Player Inventory")

def display_end_of_round_stats(chosen_player, enemies):
    print()
    print(f"\033[1mYour health: {chosen_player.health}\033[0m")
    print("\033[1mRemaining Enemies:\033[0m", end=" ")
    for enemy in enemies:
        print(f"\033[1m{enemy.name}\033[0m (Health: {enemy.health})", end=" ") 
    print()
    
def display_path_cleared(choose_path_object):
    print()
    print(f"The \033[1m{choose_path_object.name}\033[0m has been cleared!")
    print("Tired from your adventure, you return to base and rest for the night.")
    print()
    print()
    print("You awake the next morning and continue your journey.")
    print()
    print("-------------------------------------------------------------------")


    
#Functions with Player Inputs
def choose_class():
    while True:
        display_classes_with_inventory()
        choice = input("Choose a player class (1-4): ")
        print()

        if choice == '1':
            return Knight()
        elif choice == '2':
            return Wizard()
        elif choice == '3':
            return Archer()
        elif choice == '4':
            return Rogue()
        else:
            print("Invalid choice")

def choose_path(cleared_paths):
    while True:
        # Display path options
        print("\033[1mChoose your path:\033[0m")
        available_paths = {
            '1': undergroundCavern(),
            '2': abandonedCastle(),
            '3': cursedForest()
        }
        for path_num, path_obj in available_paths.items():
            if path_obj not in cleared_paths:
                print(f"({path_num}) {path_obj.name}")
        choice = input("Choose a path (1-3): ")

        if choice in available_paths:
            chosen_path = available_paths[choice]
            cleared_paths.append(chosen_path)
            return chosen_path
        else:
            print("Invalid choice")



def choose_action(chosen_player, enemies):
    for enemy in enemies:
        enemy.dazed_duration = 0

    while enemies and chosen_player.can_take_action():
        display_action_choice(chosen_player)
        choice = input("What do you want to do? ")

        if choice == '1':  # Attack
            if chosen_player.inventory[0].boost_duration >= 0:
                chosen_player.inventory[0].boost_duration -= 1
            if enemies:
                print()
                print("Choose an enemy to attack:")
                for i, enemy in enumerate(enemies, 1):
                    print(f"{i}. {enemy.name} (Health: {enemy.health})")
                enemy_choice = input("Enter the number of the enemy to attack: ")
                if enemy_choice.isdigit():
                    enemy_choice = int(enemy_choice) - 1
                    if 0 <= enemy_choice < len(enemies):
                        enemy = enemies[enemy_choice]
                        print()
                        print(f"You attack {enemy.name} with {chosen_player.inventory[0].name} for {chosen_player.inventory[0].damage} damage!")
                        enemy.dazed_duration = 1
                        enemy_defeated = enemy.take_damage(chosen_player.inventory[0].damage)
                        chosen_player.mana -= 10
                        chosen_player.reduce_stamina()
                        if enemy_defeated:
                            enemies.remove(enemy)
                        else:
                            print(f"{enemy.name} remaining health: {enemy.health}")
                    else:
                        print("Invalid input. Please enter a number.")
                else:
                    break

        elif choice == '2':  # Block
            if chosen_player.inventory[1].cooldown <= 0:
                if enemies:
                    print("Choose an enemy to block:")
                    for i, enemy in enumerate(enemies, 1):
                        print(f"{i}. {enemy.name} (Damage output: {enemy.damage})")
                    enemy_choice = input("Enter the number of the enemy to block: ")
                    if enemy_choice.isdigit():
                        enemy_choice = int(enemy_choice) - 1
                        if 0 <= enemy_choice < len(enemies):
                            blocked_enemy = enemies[enemy_choice]
                            print(f"You block the {blocked_enemy.name} using your {chosen_player.inventory[1].name}!")
                            print(f"{blocked_enemy.name} is now dazed for two rounds!")
                            chosen_player.reduce_stamina()
                            blocked_enemy.dazed_duration = 2
                            chosen_player.inventory[1].cooldown = chosen_player.inventory[1].cooldown_duration
                            for enemy in enemies:
                                if enemy != blocked_enemy:
                                    enemy.deal_damage(chosen_player)
                        else:
                            print("Invalid input. Please enter a number.")
                    else:
                        break
                else:
                    break
            else:
                print("Shield is on cooldown. Choose another action.")

        elif choice == '3':  # Use Potion
            print("Choose a potion to use:")
            for i, potion in enumerate(chosen_player.inventory, 1):
                if isinstance(potion, Potions):
                    print(f"{i}. {potion.name}")
            potion_choice = input("Enter the number of the potion to use: ")
            if potion_choice.isdigit():
                potion_choice = int(potion_choice) - 1
                if 0 <= potion_choice < len(chosen_player.inventory):
                    potion = chosen_player.inventory[potion_choice]
                    if isinstance(potion, DamagePotion):
                        chosen_player.inventory[0].damage_boost(potion.damage_boost, potion.damage_boost_duration)
                        print(f"You used a {potion.name} and increased your damage output by {potion.damage_boost} for {potion.damage_boost_duration} rounds.")
                        chosen_player.inventory.remove(potion)
                    if isinstance(potion, PoisonPotion):
                        chosen_player.inventory[0].poison_damage(potion.poison_damage, potion.poison_damage_duration)
                        print(f"You used a {potion.name} and increased your damage output by {potion.poison_damage} for {potion.poison_damage_duration} rounds.")
                        chosen_player.inventory.remove(potion)
                    # Add handling for mana
                else:
                    print("Invalid potion choice.")
            else:
                print("Invalid input. Please enter a number.")

        elif choice == '4':
            # Craft potion
            pass
        elif choice == '5':
            display_chosen_class(chosen_player)
        else:
            print("Invalid Choice")

    if chosen_player.is_stamina_zero():
        print("Your stamina is depleted! All enemies attack you for double damage!")
        for enemy in enemies:
            enemy.deal_damage(chosen_player)
        chosen_player.stamina = 100  # Restore stamina after enemies attack

    for enemy in enemies:
        if enemy.dazed_duration <= 0:
            enemy.deal_damage(chosen_player)
        elif enemy.dazed_duration > 0:
            print(f"{enemy.name} is dazed!")
            enemy.dazed_duration -= 1

    chosen_player.inventory[0].update_boost_duration()
    display_end_of_round_stats(chosen_player, enemies)

    if enemies:
        for enemy in enemies:
            if enemy.dazed_duration <= 0:
                enemy.deal_damage(chosen_player)
            elif enemy.dazed_duration > 0:
                print(f"{enemy.name} is dazed!")
                enemy.dazed_duration -= 1
            if chosen_player.health <= 0:  # Check if player is dead
                print("You have been defeated!")
                return
        if not enemies:
            print("All enemies defeated!")
            return
    
    
def main():
    # Set Player class
    chosen_player = choose_class()
    if chosen_player: 
        # Display class selection with inventory
        display_chosen_class(chosen_player)
        
        cleared_paths = []
        all_paths_cleared = False
        
        while not all_paths_cleared:
            # Choose path to enter
            choose_path_object = choose_path(cleared_paths)
            if choose_path_object:
                enemies = choose_path_object.inventory
                # Display enemies in chosen path
                display_chosen_path(choose_path_object, enemies)
                
                while enemies:
                    # Choose player action
                    enemies = choose_action(chosen_player, enemies)
                    if not enemies:   # Round winning conditions
                        display_path_cleared(choose_path_object)
                        cleared_paths.append(choose_path_object)  # Add cleared path to the list
                        chosen_player.upgrade_inventory()
                        break
        if len(cleared_paths) == 3:
            print("You Win!")
        
            
if __name__ == "__main__":
    main()
```
